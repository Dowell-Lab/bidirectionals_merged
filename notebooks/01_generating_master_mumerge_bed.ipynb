# Analysis introduction and summary of findings

## Experimental setup

**Goal:** Assess the muMerge files by paper and cell type for `dREG` and `Tfit` and produce `Master muMerge` annotations.

**Input:** 

1. muMerge file where each paper was mumerged by paper condition then subsequently muMerged by cell-type/tissue type. The muMerging was done for Tfit and dREG samples separately. This was done for human (hg38) and mouse (mm10) samples.

    - All_tiers_dreg_mm10_MUMERGE.bed
    
    - All_tiers_tfit_mm10_MUMERGE.bed
    
    - All_tiers_dreg_hg38_MUMERGE.bed
    
    - All_tiers_tfit_hg38_MUMERGE.bed

**Analysis workflow:**

1. In each bidirectional caller mumerge, the papers that support a call are noted

2. The first step is to count how many papers support a bidirectional call

3. Then we assess if these calls are also called by both `Tfit` and `dREG`

The ultimate goal of this analysis is to backtrack whether a *singleton (i.e. a call in only one paper)* is also a singleton in that said paper *(i.e. called in a single sample)*.

**Output:** 

Master bidirectional bed6 files for Mouse (mm10) and Human (hg38) genomes. 

When `dREG` and `Tfit` overlap, the region is kept. Additionally, singletons were only kept if they are from a paper with QC 1 or 2 and GC greater than 0.49. Lastly, regions with lengths <150bp & >2kb (mouse) >2.5kb (human) were removed. 

## Results

**Takeaways:**

1. About **40-45%** of calls for both dREG and Tfit are singletons (i.e. appear in a single paper)

2. Approximately **25-28%** are _TRUE_ singletons when considering overlap between dREG and Tfit calls.

3. Surprisingly, **<%10 (52,556 regions in Mouse & 53,763 regions in Human)** of overlapping calls have a matched paper support. Meaning, the same papers support the call in both Tfit AND dREG.

4. Tfit tends to call longer regions that are mostly genes. Therefor, large regions (>2.5kb in humans and >2kb in mouse) were removed. 

5. We also filtered calls that were shorted than the longest read lenth (150bp), as these regions are usually called due to reads piling up.

6. Filtering regions with low QC, low GC and by length did not drastically affect the recovery of FANTOM5 bidirectionals.

````r
##from https://stackoverflow.com/questions/35511951/r-ggplot2-collapse-or-remove-segment-of-y-axis-from-scatter-plot
squish_trans <- function(from, to, factor) {
  
  trans <- function(x) {
    
    if (any(is.na(x))) return(x)

    # get indices for the relevant regions
    isq <- x > from & x < to
    ito <- x >= to
    
    # apply transformation
    x[isq] <- from + (x[isq] - from)/factor
    x[ito] <- from + (to - from)/factor + (x[ito] - to)
    
    return(x)
  }

  inv <- function(x) {
    
    if (any(is.na(x))) return(x)

    # get indices for the relevant regions
    isq <- x > from & x < from + (to - from)/factor
    ito <- x >= from + (to - from)/factor
    
    # apply transformation
    x[isq] <- from + (x[isq] - from) * factor
    x[ito] <- to + (x[ito] - (from + (to - from)/factor))
    
    return(x)
  }
  
  # return the transformation
  return(trans_new("squished", trans, inv))
}
````

mumerge=${wd}/processed_data/master_mumerge/exploratory

for bed in `ls ${mumerge} | grep "mm10"`; do

    bedtools sort -i ${mumerge}/${bed} | bedtools intersect -a - -b ${mm10_fantom} -wa -u > ${out}/bidir_${bed}

done 

for bed in `ls ${mumerge} | grep "hg38"`; do

    bedtools sort -i ${mumerge}/${bed} | bedtools intersect -a ${hg38_fantom} -b - -wa > ${out}/F5_${bed}
    bedtools sort -i ${mumerge}/${bed} | bedtools intersect -a - -b ${hg38_fantom} -wa -u > ${out}/bidir_${bed}

done
